```python
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="my_user_agent")
city ="London"
country ="Uk"
loc = geolocator.geocode(city+','+ country)
print("latitude is :-" ,loc.latitude,"\nlongtitude is:" ,loc.longitude)
```

```python
lat = []
long = []
for item in data['City']:
    city = item
    country = 'USA'
    loc = geolocator.geocode(f'{city}, {country}')
    if loc.latitude not in lat:
        lat.append(loc.latitude)
    else:
        lat.append(loc.latitude+0.02)
    if loc.longitude not in long:
        long.append(loc.longitude)
    else:
        long.append(loc.longitude+0.02)
```

```python
data['lat'] = pd.Series(lat)

data['long'] = pd.Series(long)
```

```python
data
```

```python
import plotly.express as px
```

```python
df = data[['SB','Winner', 'Loser', 'lat', 'long', 'League']]
fig = px.scatter_mapbox(df, lat='lat', lon='long', hover_name='Winner',
                        hover_data=['SB', 'Loser'],
                        color='League',
                        zoom=3.5, height=200)

fig.update_traces(marker_size=12, selector=dict(type='scattermapbox'))

fig.update_layout(mapbox_style='open-street-map')

fig.update_layout(height=500, margin={'r':15, 't':0, 'l':15, 'b':0})
fig.show()
```
